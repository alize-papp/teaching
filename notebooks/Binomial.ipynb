```python
import numpy as np
import matplotlib.pyplot as plt
```

## Examples of random variables that follow a binomial law.

Find a random variable that follow a binomial law in each of the following cases: 
1. A pharmaceutical company conducts a clinical trial to test the efficacy of a new drug. They randomly assign 1000 patients to receive either the drug or a placebo and record whether each patient experiences a positive outcome. X: number of healed patients.
1. A company wants to analyze the conversion rate of its website, which is the proportion of visitors who make a purchase. They randomly select 500 visitors and record whether each visitor made a purchase or not. X: the number of visitors who made a purchase.
1. A survey is conducted with a sample of 200 people, and each person is asked whether they like ice-cream. 60% of people like ice-cream. X: the number of people in the sample who like ice-cream.
1. A production line manufactures a large number of items, and a quality control team randomly selects 500 items to check for defects. They record whether each item is defective or not. X: number of defective items.
1. A local government is interested in assessing the recycling behavior of households in a community. They randomly select 500 households and record whether each household actively participates in recycling or not. X: number of households who participate in recycling.

## Simulation of one binomial variable

```python
# n = 1, just one trial
# Set the parameters
n = 1  # Number of trials
p = 0.8  # Probability of success
size = 1 # Number of random draws

# Generate random draws from the binomial distribution
random_draws = np.random.binomial(n, p, size)
print(random_draws)
```

```python
# n = 10, 10 trials
# Set the parameters
n = 10  # Number of trials
p = 0.8  # Probability of success
size = 1 # Number of random draws

# Generate random draws from the binomial distribution
random_draws = np.random.binomial(n, p, size)
print(random_draws)
```

## Simulations of the distribution

```python
# Set the parameters
n = 10  # Number of trials
p = 0.8  # Probability of success
size = 1000 # Number of random draws

# Generate random draws from the binomial distribution
random_draws = np.random.binomial(n, p, size)
#random_draws

# Plot the distribution
plt.hist(random_draws, bins='auto', alpha=0.7, rwidth=0.85)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution')
plt.grid(True)
plt.show()
```

```python
# Set the parameters
n = 10  # Number of trials
p = 0.5  # Probability of success
size = 1000  # Number of random draws

# Initialize an empty list to store the evolving distribution
evolving_distribution = []

# Draw samples incrementally and update the evolving distribution
increment = 1

for i in range(1, size+1):
    
    random_draw = np.random.binomial(n, p)
    evolving_distribution.append(random_draw)

    # Plot the evolving distribution
    if i % increment == 0:
        print(i, increment, len(evolving_distribution))
        
        plt.clf()  # Clear the previous plot
        plt.hist(evolving_distribution, bins=np.arange(n+2)-0.5, alpha=0.7, rwidth=0.85)
        plt.xlabel('Number of Successes')
        plt.ylabel('Frequency')
        plt.title('Evolving Binomial Distribution (Size: {})'.format(i))
        plt.grid(True)
        plt.pause(0.01)  # Pause to show the evolving plot
    
    # Plot 1 by 1 at first, then 10 by 10, then 100 by 100
    if i % 10 == 0 and i < 100:
        increment = 10
    if i % 100 == 0:
        increment = 100


plt.show()
```

## Other distributions

```python
size = 1000
# Generate samples for each distribution
n = 10  # Number of trials for Binomial distribution
p = 0.3  # Probability of success for Binomial distribution
binomial_samples = np.random.binomial(n, p, size=size)

p = 0.3  # Probability of success for Geometric distribution
geometric_samples = np.random.geometric(p, size=size)

lam = 2  # Lambda parameter for Poisson distribution
poisson_samples = np.random.poisson(lam, size=size)

low = 1  # Lower bound for Uniform distribution
high = 7  # Upper bound for Uniform distribution
uniform_samples = np.random.randint(low, high+1, size=size)

# Plot each distribution
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(binomial_samples, bins=np.arange(n+2)-0.5, alpha=0.7, rwidth=0.85)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Binomial Distribution')

plt.subplot(2, 2, 2)
plt.hist(geometric_samples, bins='auto', alpha=0.7, rwidth=0.85)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Geometric Distribution')

plt.subplot(2, 2, 3)
plt.hist(poisson_samples, bins='auto', alpha=0.7, rwidth=0.85)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Poisson Distribution')

plt.subplot(2, 2, 4)
plt.hist(uniform_samples, bins=np.arange(low, high+2)-0.5, alpha=0.7, rwidth=0.85)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Uniform Distribution')

plt.tight_layout()
plt.show()
```
