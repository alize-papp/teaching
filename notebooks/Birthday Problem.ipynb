# Birthday Problem
The birthday problem is a classic probability problem. 

All about this problem: https://en.wikipedia.org/wiki/Birthday_problem

See variations of it: https://en.wikipedia.org/wiki/Birthday_problem#Other_birthday_problems

```python
import math

import numpy as np
import matplotlib.pyplot as plt
```

```python
# the probability of n people not having any of their birthday in common

def proba_birthday(n):
    p = math.factorial(365) / (math.factorial(365-n) * 365**n)
    return p

# eg: 20 people, the probability that none have the same birthday is 59%
proba_birthday(20)
```

```python
# Let's see how the probability changes as the groups groups
# x the number of people in the group
x = range(1, 365)
# y the probability of no person have the same birthday
y = []
# z the probability that at least 2 people have the same birthday
z = []

for i in x:
    p = proba_birthday(i)
    y.append(p)
    z.append(1 - p)
```

```python
print(y)
print(z)
```

```python
plt.plot(x, z)
```

```python
# For interactive charts, you can use the plotly library https://plotly.com/python/
import plotly.express as px
fig = px.line(x=x, y=z)
fig.show()
```
